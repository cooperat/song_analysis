```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cluster = pd.read_csv('data/songsclustered.csv')
artist = pd.read_csv('data/genre_DB.csv')
artist.drop('Unnamed: 0',1, inplace=True)
```

```python
full = cluster.join(artist.set_index('Artist'), on='artist')
print(full.shape)
full.head()
```

```python
full.columns
```

## Aperçu global 

```python
for i in [3, 5, 8, 10, 15, 20, 30, 50]:
    grouped = full.groupby(['cluster_'+str(i)])
    clusters_distrib = grouped.count()
    clusters_distrib['song'].plot.bar()
    plt.rcParams["figure.figsize"] = (50, 10)
    plt.show()

    #  for each cluster, display the distribution of songs
    for j in range(i):
        print('\n', 'Cluster number :', j)
        cluster = grouped.get_group(j)
        print(cluster['artist'].describe())
```
