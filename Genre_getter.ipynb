```python
import pandas as pd
import wptools
import re
```

```python
data = pd.read_csv('data/songdata.csv')
data.head()
```

```python
artists = list(data["artist"].unique())
```

```python
info = []
for art in artists : 
    try :
        
        page = wptools.page(art)
        page.get_parse()

        if page.data['infobox'] is not None :
            if 'years_active' in page.data['infobox']:
                year = page.data['infobox']['years_active']
            else:
                year = None

            genre = page.data['infobox']

            if 'origin' in page.data['infobox']: 
                origin = page.data['infobox']['origin']
            else : 
                origin = None
        else :
            genre, year, origin = None, None, None
        info.append((art, genre, year, origin))
    except : 
        info.append((art, None, None, None))
```

```python
re.sub(r"\D", "",'1991&ndash;present18902')[:8]
```

```python
full_art = []

for x in info :
    art = [0]*16
    art[0] = x[0] # Artiste
    if x[2] is not None: # Year
        art[1] = re.sub(r"\D", "", x[2])[:8]
        if len(art[1]) == 4 : art[1] = str(art[1])+"-present"
        elif len(art[1]) == 8 : art[1] = str(art[1][:4])+"-"+str(art[1][4:])
    if x[3] is not None: # Origin
        art[2] = x[3].split(",")[-1].replace("[[","").replace("]]","")
    if x[1] is not None: # Genre
        content = str(re.findall('\[\[.*?\]\]',str(x[1])))
        if "pop" in str(content).lower() : art[3] = 1
        if "blues" in str(content).lower() : art[4] = 1  
        if "classical" in str(content).lower() : art[5] = 1
        if "country" in str(content).lower() : art[6] = 1
        if "electronic" in str(content).lower() : art[7] = 1
        if "hip-hop" in str(content).lower() or "rap" in str(content).lower(): art[8] = 1
        if "jazz" in str(content).lower() : art[9] = 1
        if "r&b" in str(content).lower() : art[10] = 1
        if "reggae" in str(content).lower() : art[11] = 1
        if "rock" in str(content).lower() : art[12] = 1
        if "latin" in str(content).lower() : art[13] = 1
        if "soul" in str(content).lower() : art[14] = 1     
        if "metal" in str(content).lower() : art[15] = 1  
    full_art.append(art)
```

```python
tab = pd.DataFrame(full_art, columns=["Artist","Year","Origin","Pop","Blues","Classical","Country","Electronic","Hip-hop","Jazz","R&B","Reggae","Rock","Latin","Soul","Metal"])
```

```python
tab["no_genre"]=tab.sum(1,numeric_only=True)
```

```python
use = tab[tab["no_genre"]>0]
```

```python
app = pd.read_csv("a_trier.csv", sep=';')
app.drop('Unnamed: 0', 1, inplace=True)
app["no_genre"]=app.sum(1,numeric_only=True)
app[app["no_genre"]==0]
```

```python
full = use.append(app, ignore_index=True)
```
